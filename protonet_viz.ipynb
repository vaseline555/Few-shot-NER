```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
emb = np.random.randn(116, 2)
```

```python
proto = np.random.randn(6, 2)
```

```python
curr_label = np.random.randint(0, 6, size=(116))
```

```python
curr_pred = np.random.randint(0, 6, size=(116))
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(x=emb[:, 0], y=emb[:, 1], c=curr_label, s=1e2, marker='+', alpha=0.4, label='true')
plt.scatter(x=emb[:, 0], y=emb[:, 1], c=curr_pred, s=1e2, marker='o', alpha=0.8, label='pred')
plt.scatter(x=proto[:, 0], y=proto[:, 0], s=1e4, c=[i for i in range(proto.shape[0])], alpha=0.1)
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
for g in np.unique(curr_label):
    idx_label = np.where(curr_label == g)
    idx_pred = np.where(curr_pred == g)
    #ax.scatter(x=emb[:, 0][idx_label], y=emb[:, 1][idx_label], s=1e2, alpha=0.5, label=g)
ax.scatter(x=emb[:, 0], y=emb[:, 1], s=5e2, alpha=0.5, c=curr_label != curr_pred, marker='+') 
ax.legend()
ax.plot(proto[:, 0], proto[:, 1], 'ro', markersize=100, alpha=0.1)
for idx, (x, y) in enumerate(proto):
    ax.annotate([str(i) for i in range(len(proto))][idx],  xy=(x - 0.05, y - 0.05))
plt.show()
```
